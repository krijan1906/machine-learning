### Exception
* An exception is an unwanted or unexpected event that occurs during the execution of a program (i.e at runtime) and distrupts the normal flow of program's instructions. It occurs when something unexpected things happen, like accessing invalid index, dividing by zero, or trying to open a file that does not exist.
  
* An exception is a runtime error which can be handled by the programmer.
* All exceptions are represented as classes in python.

### Types of Exceptions:
* Built-in Exception: Exceptions which are already available in python language. 
* User Defined Exception:A programmer can create his own exceptions, called user-defined exceptions.

### Exception Handling
* Exception Handling handles errors that occur during the execution of a program.
* This allows to respond to the error, instead of crashing the running program.
* It enables you to catch and manage errors, making your code more robust and user-friendly.

### Need of Exception Handling
* When an exception occurs, the program terminates suddenly.
* Suddenly termination of program may corrupt the program.
* Exception may cause data loss from the database or a file.

### Advantages of Exception Handling:
* Improved program reliability: By handling exceptions properly, you can prevent your program from crashing or producing incorrect results due to unexpected errors or input.
* Simplified error handling: Exception handling allows you to separate error handling code from the main program logic, making it easier to read and maintain your code.
* Cleaner code: With exception handling, you can avoid using complex conditional statements to check for errors, leading to cleaner and more readable code.
* Easier debugging: When an exception is raised, the Python interpreter prints a traceback that shows the exact location where the exception occurred, making it easier to debug your code.

### We use different block to handle exceptions.

#### Try - The try block contains code which may cause exceptions.
* Syntax:
* try:
  * statements
  

#### Except - The except block is used to catch an exception that is raised/occur in the try block. There can be multiple except block for try block.
* Syntax:
* except ExceptionName:
  * statements

#### Else - This block will get executed when no exception is raised/occur. Else block is executed after try block.
* Syntax:
* else:
  * statements

#### Finally - This block will get executed irrespective of whether there is an exception or not.
* Syntax:
* finally:
  * statements

### Points to remember while handling exceptions:
* We can write several except blocks for a single try block.
* We can write multiple except blocks to handle multiple exceptions.
* We can write try block without any except blocks.
* We can not write except block without a try block.
* Finally block is always executed irrespective of whether there is an exception or not.
* Else block is optional.
* Finally block is optional.

```python
# Examples: 
'''
try:
    statement
except ExceptionClassName:
    statement
else:
    statement
finally:
    statement
'''
```

```python
# Example:
# ZeroDivisionError
a = 10
b = 0
d = a/b
print(d)
print('Rest of the code')
```

```python
# The above exception can be handle as:
a = 10
b = 0
try:
    d = a/b  # this will throws error so to handle this we use except
    print(d)
    print('Division by 0 possible')
    
except ZeroDivisionError: # This except block is executed when exception occurs
    print('Division by zero not allowed')
    
print('Rest of the code')
```

```python
a = 10
b = 5
try:
    d = a/b  # this will not throws error so no need to handle this
    print(d)
    print('Division is possible')
    
except ZeroDivisionError:  # This except block is executed when exception occurs
    print('Division by zero not allowed')
    
print('Rest of the code')
```

```python
a = 10
b = 0
# b = 5
try:
    d = a/b  # this will not throws error so no need to handle this
    print(d)
    print('Division possible')
    
except ZeroDivisionError:  # This except block is executed when exception occurs
    print('Division by zero not allowed')
    
else: # this will execute when except block is not executed
    print('Inside Else')
    
finally:   # this will execute always whether exception occur or not
    print('Inside Finally')
    
print('Rest of the code')
```

```python
# Exception handling Examples:

# try except else finally

while True:  # Infinite loop
    try:
        age = int(input('Enter your age:')) # error will occur if string like seven instead of 7 is input
        break  # If error occurs during input
    except ValueError: # this will run when error occur in try block
        print('Maybe you entered string instead of number, try again')
  
    except:
        print('Unexpected error...')

if age < 18:
  print('You can\'t play this game.')
else:
  print('You can play this game.')
    
```

```python
# else and finally clause in expection handling

while True:
  try:
    number = int(input('Enter a number : '))
  except ValueError:
    print('Please type integer !! s')
  except:
    print('Unexcepted error !!!')
  else: # when try block successfully execute then else block execute
     print(f'User input = {number}')
     break
  finally: # this will execute whether exception occurs or not
    print(f'finally blocks.........')
```

### Assert Statement
* The assert statement is useful to ensure that a given condition is True. If it is not true, it raises AssertionError.
* Syntax: assert condition, error_message

* If the condition is False then the exception by the name AssertionError is raised along with the message.
* If message is not given and the condition is False then also AssertionError is raised without message.

```python
# Assertation error:
a = 20
assert a<=10, 'Invalid Number'  # Condition false so assertion error occurs
```

### User Defined Exception:
* A programmer can create his own exceptions, called user-defined exceptions or custom exception.
* Creating exception class using exception class as a base class.

```python
# We can create our own exception by creating a sub class to built-in Exception class.
# Syntax: 
'''
class MyException(Exception):
    pass
'''
```

```python
# User Defined Exception
class BalanceException(Exception):
    pass

def checkbalance():
    money = 10000
    # withdraw = 5000
    withdraw = 10000  # This will raise error
    balance = money - withdraw
    if(balance <= 1000): # balance is remaining money after withdraw
        raise BalanceException('Insufficient Balance') # This is passed to BalanceException object as be
    print(balance)

try: 
    checkbalance()  # calling checkbalance method

except BalanceException as be:
    print(be)
```

Error vs Exception
* An exception is an error that can be handled by programmer.
* An exception which are not handled by programmer, becomes an error.
* All exceptions occur only at runtime.
* Error may occur at compile time or runtime.
* Exception occur due to situations like invalid input, missing files or network issues.

### Error vs Warning
* It is compulsory to handle all error otherwise program will not execute, while warning represents a caution and even though it is not handled, the program will execute.
