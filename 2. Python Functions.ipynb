### Function
* A function is a reusable block of code that performs a specific task.

#### Why to Use Functions?
    * Avoid code repetition
    * Improve readability
    * Easy debugging & testing
    * Divide large task into many small task so it will help you to debug code.
    * Can remove or add new feature to a function anytime.

### Function Definition
* We can define a function using def keyword followed by function name with parentheses.
* This is also called as creating a Function, Writing a Function, Defining a Function.
* While creating function we need to maintain proper indentation.
* def represents starting of function definition.
* parentheses describes this is a function not variable or other object.
* : decribes beginning of function body.
* A function runs only when we call it, function do not run on its own.

```python
# Function without argument and parameter
def add(): # defining function/function definition
  a = 10   # local variable
  b = 20   # local variable
  c = a + b  # statements
  print(c)   # statements
    
# Calling a function without argument
add() # Calling function transfers the control to function definition part.
```

```python
def add(a, b):  # a and b are parameter
  c = a + b
  print(c)
add(20,30) # 20 and 30 are argument
```

#### Write once and use it as many time as you need.

```python
# defining a function
def disp():
  name = "Rahul"
  print("Hello", name)

# Calling a function
disp()
disp() # calling same function next time
disp() # calling same function next time
```

#### Divide large task into many small task which make easy for debuging code

```python
# Defining Function with argument and parameter
def add(y):
  x = 10
  c = x + y
  print(c)

add(20.56)         # calling a function with argument

def sub(y):
  x = 20
  c = x - y
  print(c)

sub(10)          # calling a function with argument

def mul(y):
  x = 20
  c = x * y
  print(c)

mul(10)         # calling a function with argument

def div(y):
  x = 20
  c = x / y
  print(c)

div(10)         # calling a function with argument
```

### Return Statement
* Used to return something from the function. In Python, it is possible to return one or more variables/values.

```python
# Return statement with single value
def add():
  x = 10 # local variable
  y = 20 # local variable
  c = x + y
  return c # return where function is called

# Calling a function
sum = add() # sum variable holds the value of c
print(sum)
```

```python
def add(y):
  x = 10
  c = x + y
  return c  # return where add() function is called
sum = add(20) # calling function and passing value for y
print(sum)
```

```python
 # Return statement with multiple value
def add_sub(y):
   x = 10
   c = x + y
   d = y - x
   return c, d  # return value of c and d where function is called

# Calling a function
sum, sub = add_sub(20) # sum, sub = 30, 10
print(sum)
print(sub)
```

### print vs return
#### print:
* Displays output to the screen
* Does not send data back to the caller
* Returns None
* print only shows the value, it does not store or return it
  
#### return: 
* Sends a value back from a function
* Allows value to be stored or reused

### When to use?
* Use print() when you want to display a value.
* Use return when you want to use the value later in code.

```python
# print
def add(a, b):
    print(a + b)

result = add(2, 3)
print(result)
```

```python
# return 
def add(a, b):
    return a + b

result = add(2, 3)
print(result)
```

### Nested Function:
* When we define one function inside another function, it is known as nested function or function nesting.
* Execution of function starts when function is called.

```python
# Defining function
def disp(): # This runs at 2nd
  def show():  # This runs  at 5th
    print("Show Function") # This runs at last
  print("Disp Function")  # This runs at 3rd

 # Calling function
  show()  # This runs 4rd
disp() # This runs first and control transfer/pass to disp() function at top where it is define
```

```python
# Nested function with return statement
def disp(): # 2nd
  def show(): # 4th
    return("Show Function") # 5th, return where show() is called
  result = show() + " Disp Function" # 3rd time show() called and control transfer to show() function
  return result # return result where disp() function is called

 # Calling function
a = disp() # 1st execute and control transfer to disp() function at top ; a store the value of result
print(a)  # Print the value stored at a
```

### Actual and Formal argument
* Formal argument - Function definition parameters are called as formal arguments.
* Actual argument - Function call arguments are actual arguments.

```python
def add(x, y): # Formal arguments rather than calling parameter
  c = x + y
  print(c)

add(10, 20) # Actual arguments
```

### Types of actual arguments
* Positional arguments
* Keyword arguments
* Default arguments
* Variable arguments
* Keyword Variable Length arguments

### Positional arguments
* These arguments are passed to the function in correct position order.
* The number of arguments and their positions in the function definition should be equal to the number and position of the argument in the function call.

```python
# Positional arguments example
def pw(x, y): # this function has 2 arguments
  z = x ** y
  print(z)

pw(5, 2) # x = 5, y = 2, order/position should be maintained
```

### Keyword arguments
* These arguments are passed to the function with name-value pair so keyword arguments can identify the formal argument by their names.
* The keyword argument's name and formal argument's name must match.


```python
# Keyword arguments example
def show(name, age):
         print(name, age)

show(name ="Rahul", age = 30) # no need to maintain order but name of argument must be same
```

### Default arguments
* Sometimes we mention default value to the formal argument in function definition and we may not required to provide actual argument, In this case default argument will be used by formal argument.
* If we do not provide actual argument for formal argument explicitly while calling the function then formal argument will use default value on the other hand if we provide actual argument then it will use provided value.

```python
# Default argument example
def show(name, age=27): # age default value 27
  print(name, age)

show(name="Rahul") # age is not passed so takes default value of age
```

### Variable length arguments
* Variable length argument is an argument that can accept any number of values.
* The variable length argument is written with * symbol.
* It stores all the value in a tuple.
* We can pass any number of arguments using variable length arguments.
* We can use index number to access variable length arguments.

```python
# Normal function
def add(x, y):  # this function takes 2 arguments
  z = x + y
  print("Addition", z)
add(5, 2) # Here, we cannot pass more than 2 value
```

```python
# Variable length arguments
def add(*num): # takes any number of arguments
    print(num[0]) # accessing 1st value
    print(num[1]) # accessing 2nd value
    print(num[2]) # accessing 3rd value
add(5, 2, 4) # passing 3 arguments
```

```python
# Variable length arguments
def add(*num):  # * means variable length arguments, here we can pass any number of arguments
  z = num[0] + num[1] + num[2]  # accessing value using index
  print("Addition", z)

add(5, 2, 4) # Here, we can pass any number of values
```

```python
# Variable length arguments
def add(x, *num):
  z = x + num[0] + num[1]
  print("Addition", z)

add(5, 2, 4) # Here, we can pass any number of values
```

### Keyword variable length arguments
* Argument that can accept any number of values provided in the form of key-value-pair.
* The keyword variable length argument is written with ** symbol
* It stores all the value in a dictionary in the form of key-value-pair.

```python
# Keyword Variable length arguments
def add(**num):
  z = num['a'] + num['b'] + num['c']  # accessing values through key
  print("Addition", z)

add(a=5, b=2, c=4) # passing values in key-value pair
```

```python
# Variable length arguments
def add(x, **num): # x is normal argument
  z = x + num['a'] + num['b'] # only using 2 value
  print("Addition", z)

add(3, a=5, b=2, c=12)  # passing 3 values in key-value pair, do not occur error
```

### Anonymous Function or Lambdas
* A function without a name is called as anonymous function.
* It is also known as Lambda Function.
* Anonymous function are not defined using def keyword rather they are defined using lambda keyword.
*  Syntax: lambda argument_list: expression

```python
# Creating a lambda function
lambda x,y : x + y
# lambda keyword represents an anonymous function is being created.
# x, y is an argument list.
# : represents beginning of the function.
# x + y is an expression

# Calling lambda function
sum = lambda x : x + 1  # lambda function return a function which is stored at sum 
sum(5) # calling function using sum() and passing 5 as argument for x
```

```python
# Calling lambda function
add = lambda x, y : x + y
add(5,2)   # calling add function and passing 5 and  2 arguments for x and y
```

* Lambda function doesn't have any name. eg. lambda x : print(x)
* Lambda function returns a function which is assigned using function variable. eg. show = lambda x : print(x)
* We can call this function using that function variable.
* Lambda function can take zero or any number of argument but returns/contains only one expression. eg. lambda x, y : x + y
* In lambda function there is no need to write return statement.
* It can only contain expressions and can't include statements in its body. i.e we cannot use if statement, for loop etc.
* You can use all the type of actual arguments. eg. positional argument, keyword argument, variable length argument etc.

```python
# Normal function
def show(x):
    print(x)
show(5)
```

```python
# The above task can be done using Anonymous Function or lambda Function
show = lambda x : print(x)     # show is function variable
show(5)  # calling function and passing value for x
```

```python
# Passing single values
add = lambda x, y : x + y # return value of x+y where function is called
a = add(5,2)  # calling function will transfer control to 1st line  
print(a)
```

```python
# Passing multiple values
add_sub = lambda x, y : (x + y, x - y)
a, s = add_sub(5,2)  # calling function will transfer control to 1st line  
print(a)
print(s)
```

```python
# Passing default argument
add = lambda x, y=3 : x + y
print(add(5))
```

### Local Variables
* Variable which are declared inside a function called as local variable.
* Local variable scope is limited only to that function where it is created.
* Local variable value is available only in that function not outside of that function.

```python
# Local variable
def show():
  x = 10 # local variable inside show function
  print(x)

show()

# we cannot access local variable outside the scope will show error -->NameError: name 'x' is not defined
print(x)
```

```python
def add(y):
  x = 10  # local variable
  print(x)  # using local variable inside function
  print(x + y)  # using local variable inside function
add(20) # calling function

print(x) # using local variable outside function, it will show error -->NameError: name 'x' is not defined
```

### Global variables
* When a variable is declared above a function, it becomes global variable.
* These variables are available to all the function which are written after it.
* The scope of global variable is the entire program body written below it.

```python
a = 50 # Global varaible

def show():
  x = 10    # local variable
  print(x)  # using local variable inside function
  print(a)  # using global variable inside function

show()      # calling function

# print("Local Variable:", x)  # using local variable outside function, it will show error.
print("Global Variable:", a)   # using global variable outside function
```
