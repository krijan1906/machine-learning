### Class Method
* class methods are the methods which act upon the class variables or static variable of the class.
* Decorator @classmethod need to write above the class method.
* By default, the first parameter of class method is cls which refers to the class itself.

```python
# class method without parameter
class Mobile:
      fp = 'Yes'  # class variable
    
      @classmethod        # Decorator
      def show_model(cls):   # class method without parameter
        print("Fingerprint:", cls.fp) # Accessing class variable

realme = Mobile()
realme.show_model()   # calling class method without argument
```

```python
### Calling class Method with Argument
# Syntax:
'''
Classname.method_name(actual_argument)
'''
```

```python
# Example:
# class method with parameter
class Mobile:
      fp = 'Yes'      # class variable
      
      @classmethod        # Decorator
      def show_model(cls, r):   # defining class method with parameter
        cls.ram = r             # Assigning r to variable ram
        print("Fingerprint:", cls.fp)  # Accessing class variable inside class method
        print("RAM:", cls.ram)

realme = Mobile()    # creating object
Mobile.show_model("4GB")   # calling class method with argument
```

### Static Methods
* Used when some processing is related to the class but does not need the class or its instance to perform any work.
* We use static method when we want to pass some values from outside and perform some action in the method.
* Decorator @staticmethod need to write above the static method.

```python
# Static Method without parameter
class Mobile:
  @staticmethod  # Decorator
  def show_model():   # Static method
    print("RealMe X")

realme = Mobile()   # Creating object of Mobile class
Mobile.show_model()   # Calling static method without argument
```

```python
# Static Method with parameter
class Mobile:
  @staticmethod  # Decorator
  def show_model(m, p):   # Defining method with parameter
    model = m
    price = p
    print(model, price)

realme = Mobile()
Mobile.show_model('RealMe X', 1000)   # Calling static method with argument and passing values to m and p
```

```python
# Passing Members of one class to another class
class Student:
  # Constructor
  def __init__(self, n, r):
    self.name = n  # Assigning n to self.name
    self.roll = r  # Assigning m to self.roll

  # Instance method
  def disp(self):
    print("Stduent Name:", self.name)
    print("Student Roll:", self.roll)

class User:
  # Static method
  @staticmethod
  def show(s):   # s takes stu
    print("User Name:", s.name)
    print("User Roll:", s.roll)
    s.disp()    # calling disp() 

# Creating object of Student class
stu = Student('Rahul', 101)  # Passing values to n=Rahul and m=101

# Calling static method of User class
User.show(stu)  # passing stu object to User's class
```

### Nested Class
* A class within a class is called as nested class or nesting of a class.
* Inner class object is created inside outer class constructor.

```python
class Student:  # Outer class
    def __init__(self, name):
        self.name = name
        self.laptop = self.Laptop()   ## creating inner class object

    def show(self):
        print("Student Name:", self.name)
        self.laptop.show()  # calling inner class show method

    class Laptop: # Inner class
        def __init__(self):
            self.brand = "Dell"
            self.ram = "16GB"

        def show(self):
            print("Laptop Brand:", self.brand)
            print("RAM:", self.ram)

# Create object of outer class
s1 = Student("Rahul")
s1.show()
```

```python
# Nested Class
# A class within a class is called as nested class or nesting of a class.
class Army:               # Outer class
      def __init__(self):  # Constructor method
        self.name = 'Rahul'  # Instance variable
        self.gn = self.Gun()  # Creating inner class object
    
      def show(self):        # Outer class method
        print("Name:", self.name)
    
      class Gun:             # Inner class
        def __init__(self):  # constructor
          self.name ='AK47'
          self.capacity = '75 Rounds'  # Instance variable
          self.length = '34.3 in'     # Instance variable
    
        def disp(self):       # Inner class method
          print("Gun Name:", self.name)
          print("Capacity:", self.capacity)
          print("Length:", self.length)

a = Army()  # create object of outer class
print(a.name)    # print Rahul
a.show()      # calling outerclass show() method

#print(a.gn.name)
#g = a.gn
# this can be also be written as:
g = Army().Gun()  # Creating inner class object

print()

print(g.name)       # Accessing inner class name
print(g.capacity)
print(g.length)
print()

g.disp()  # calling inner class method
```

```python
# Nested Class Example
# A class inside another class is called a nested class

class Army:                     # Outer class
    def __init__(self):
        self.name = "Rahul"      # Army person name
        self.gun = self.Gun()    # Creating object of inner class

    def show(self):              # Outer class method
        print("Name:", self.name)

    class Gun:                   # Inner class
        def __init__(self):
            self.name = "AK47"
            self.capacity = "75 Rounds"
            self.length = "34.3 in"

        def display(self):        # Inner class method
            print("Gun Name:", self.name)
            print("Capacity:", self.capacity)
            print("Length:", self.length)

# Create object of outer class
a = Army()
a.show()            # Call outer class method

# Access inner class object through outer class
a.gun.display()

# Create inner class object directly
g = Army.Gun()
g.display()
```
