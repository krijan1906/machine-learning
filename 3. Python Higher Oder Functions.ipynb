### Higher Order Function
Higher-order functions treat functions just like variables.

Functions that can take another function as an argument or return a function.
  
* You can pass them
* You can return them
* You can store them

### Types:
    * filter() function
    * map() function
    * reduce() function
  
filter() function - used to filter out the elements of an iterable (sequence) depending on function that tests each element in the sequence to be true or not.
* It returns those elements of sequence, for which function is true that means Select elements based on condition.


* Syntax: filter(function_name, iterable)

* function_name - it's name of a function which tests each element in the sequence return True or False.
* If function is None, returns the elements that are true.
* iterable - iterable may be either a sequence, list, string, tuple, a container which supports iteration, or an iterator.

```python
a = [10, 50, 60, 80, 90, 5, 45, 65]  # list

# We need to create a function which returns marks higher than 60

def high_marks(n):
  if n>=60:
    return True # If condition is true then return that element from list

# filter the marks >= 60
result = filter(high_marks, a) # this will return filter object; a is parameter which is list
print(result)    # this will return filter object
print(type(result))

for i in result:
  print(i)
```

```python
a = [10, 50, 60, 80, 90, 5, 45, 65]

def high_marks(n):
  if n>=60:
    return True
      
# calling filter function
result = list(filter(high_marks, a))  # converting filter object to list
print(result)   # this will return list
print(type(result))

for i in result:
  print(i)
```

```python
# The above filter function can be enhanced using lambda function or lambda expression

a = [10, 50, 60, 80, 90, 5, 45, 65]

# converting filter to list using lambda function
result = list(filter(lambda n: (n>=60), a))
print(result)
print(type(result))

for i in result:
  print(i)
```

```python
# Filter function

# list with True, False value
a = [True, False, False, True, False, True, True]  # True, False is case-sensitive

# converting filter to list using lambda function
result = list(filter(None, a))  # this will return True only
print(result)
print(type(result))

for i in result:
  print(i)
```

map() function - This function executes a specified function on each element of the iterable (sequence) and perhaps changes the elements.

* Applies a function to each item in an iterable.
* Syntax: map(function_name, iterable)
  
* function_name - it's name of a function which perform an operation on all the elements of the sequence and modified elements are returned which can be stored in another sequence.
* iterable - iterable may be either  a sequence, list, string, tuple, a container which supports iteration, or an iterator.

```python
a = [10, 20, 30, 40, 50] # list

# We need a result with each element + 2

def inc(n): 
  return n + 2

result = map(inc,a)
print(result)   # this will return map object
print(type(result))

for i in result:
  print(i)
```

```python
a = [10, 20, 30, 40, 50] # list

# We need a result with each element + 2

def inc(n):  # n can be 10, 20, 30, 40, 50
  return n + 2

# calling map function
result = list(map(inc,a))  # Converting map object to list
print(result)   # this will return map object
print(type(result))

for i in result:
  print(i)
```

```python
# The above task can also be done using lambda function also.

a = [10, 20, 30, 40, 50]

result = list(map(lambda n : n+2, a))
print(result)
print(type(result))

for i in result:
  print(i)
```

```python
# The above task can also be done using lambda function also.

a = [10, 20, 30, 40, 50]

result = map(lambda n : n+2, a)
result1 = list(result)
print(result1)
print(type(result1))

# for i in result:
  # print(i)
```

```python
# Using lambda function

a = [10, 20, 30, 40, 50]   # list
b = [1, 2, 3, 4, 5]  # list

# Here, the size of each list should be same
result = list(map(lambda n, m : n+m, a, b))  # n represents a and m represents b; the value of n+m is stored at result
print(result)
print(type(result))

for i in result:
  print(i)
```

reduce() function - This function is used to reduce a sequence of elements to a single value by processing the elements according to a function supplied.

* It returns a single value.
* This function is a part of functools module so you have to import it before using this means reduce() is not built-in.
* Syntax: from functools import reduce
* reduce(function_name, sequence)

```python
from functools import reduce
a = [10, 20, 30, 40, 50]   # list


result = reduce(lambda n, m : n+m, a)  # this will return single value; 1st time n=10, m=20, 2nd time n=30, m=30, 3rd time n=60, m=40, 4th time n=100, m=50, finally result = 
print(result)
print(type(result))
```
