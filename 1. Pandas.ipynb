### Accessing element of Series:
#### Position-based Indexing
* Use the index operator []to access an element in a series. The index must be an integer.
* In order to access multiple elements from a series we use Slice operation.

#### Label-based Indexing
* In order to access an element from series, we have to set values by index label. 
* A Series is like a fixed-size dictionary in that we can get and set values by index label.

```python
# Position-based indexing
import pandas as pd
import numpy as np

data = np.array(['a','p','p', 'l', 'e'])
s = pd.Series(data)

print(s[:3]) # returns value from index 0 to index 2
```

```python
# Label-based Indexing
import pandas as pd
import numpy as np

data = np.array(['a','p','p', 'l', 'e']) # nd array

s = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])  # Changing index
print(s)
print()
print(s[3]) # return index 3 value
```

```python
# Floating data
x = [3,4,5,6]   # list
s = pd.Series(x, index=['a', 's', 'd', 'f'], dtype="float")
print(s)
```

### Indexing and Selecting Data in Series:
* Indexing in pandas means simply selecting particular data from a Series.
* We can use .iloc[] for position-based selection and .loc[] for label-based selection.

```python
# Indexing a Series using .loc[]

import pandas as pd
import numpy as np

data = np.array(['a','p','p', 'l', 'e'])

s = pd.Series(data)

# We access the element of series using .loc[] function.
# Since loc[] includes both start and end labels
s.loc[1:3] # return from index 1 to 3
```

```python
# Indexing a Series using .iloc[]
# .iloc[] function allows us to retrieve data by position
# The df.iloc indexer is very similar to df.loc but only uses integer locations to make its selections.

import pandas as pd
import numpy as np

data = np.array(['a','p','p', 'l', 'e'])

s = pd.Series(data)
# We access the element of Series using .iloc[] function.
s.iloc[1:3]
```

We used iloc [1:3] which works on position-based indexing like Python lists. 

It fetch elements at index positions 1 to 2 excluding index 3.

### Binary Operations on Pandas Series:
* Pandas allows performing binary operations on Series, such as addition, subtraction, multiplication and division.
* These operations can be performed using functions like .add() , .sub(), .mul() and .div().


```python
# Performing Binary Operations
import pandas as pd

s1 = pd.Series([1, 2, 3], index=['A', 'B', 'C'])
s2 = pd.Series([4, 5, 6], index=['A', 'B', 'C'])

df_sum = s1.add(s2)
print(df_sum)
```

 Here, Each element of series 1 is added to the corresponding element of series 2 by matching the index labels (A,B,C).

```python
#  Selecting Specific Rows and Columns
# We can select specific rows and columns by providing lists of row labels and column names.
# Syntax: Dataframe.loc[["row1", "row2"], ["column1", "column2", "column3"]]

import pandas as pd
data = pd.read_csv('example.csv')
d1 = data.loc[[1, 2],['first_name', 'last_name', 'age']]
print(d1)
```

```python
# Selecting All Rows and Specific Columns
# We can select all rows and specific columns by using a colon [:] to indicate all rows followed by the list of column names.
# Syntax: Dataframe.loc[:, ["column1", "column2", "column3"]]

d1 = data.loc[:, ['first_name', 'last_name', 'age']]
print(d1)
```

### Slicing Pandas Dataframe
* Slicing a Pandas DataFrame is a important skill for extracting specific data subsets. Whether you want to select rows, columns or individual cells.
* Pandas provides efficient methods like iloc[] and loc[]. 
* We can use integer-based and label-based indexing to slice DataFrames

```python
# Create a Custom Dataframe
import pandas as pd

player_list = [['M.S.Dhoni', 36, 75, 5428000],
               ['A.B.D Villers', 38, 74, 3428000],
               ['V.Kohli', 31, 70, 8428000],
               ['S.Smith', 34, 80, 4428000],
               ['C.Gayle', 40, 100, 4528000],
               ['J.Root', 33, 72, 7028000],
               ['K.Peterson', 42, 85, 2528000]]

df = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'])
print(df)
```

```python
# Slicing Using iloc[] (Integer-Based Indexing)
# This allows us to extract specific rows and columns based on their integer positions starting from 0.

# Slicing Rows in dataframe:
# Row slicing means selecting a specific set of rows from the DataFrame and keeping all columns.
df1 = df.iloc[0:4]
print(df1)
```

```python
# Slicing Columns in dataframe
# Column slicing means selecting a specific set of columns from the DataFrame while keeping all rows.
df1 = df.iloc[:, 0:2]
print(df1)
```

```python
# Selecting a Specific Cell in a Pandas DataFrame
# If you need a single value from a DataFrame you can specify the exact row and column position.
value = df.iloc[3, 2]  # row 3 and col 2
print("Specific Cell Value:", value)
```

```python
# Slicing Using loc[]
# We can also slice using the loc function.
# Slicing Rows in Dataframe:
# With loc[] we can extract a range of rows by their labels instead of integer positions.

data = df.loc['M.S.Dhoni':'S.Smith']
print(data)
```

### Merging, Joining and Concatenating
##### Concatenating DataFrame using .concat()
* To concatenate DataFrames, we use the pd.concat() function.

```python
import pandas as pd

data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}

data2 = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']}

df1 = pd.DataFrame(data1, index=[0, 1, 2, 3])

df2 = pd.DataFrame(data2, index=[4, 5, 6, 7])

print(df1, "\n\n", df2)
```

```python
# We apply .concat function in order to concat two dataframe.
frames = [df1, df2]

result = pd.concat(frames)
result
```

### Concatenating DataFrames by Setting Logic on Axes:
* We can modify the concatenation by setting logic on the axes. Specifically we can choose whether to take the Union (join='outer') or Intersection (join='inner') of columns.
* We can also set axes join = inner for intersection of dataframe which keeps only the common columns.

```python
   
import pandas as pd

data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']
         }

data2 = {'Name': ['Gaurav', 'Anuj', 'Dhiraj', 'Hitesh'],
         'Age': [22, 32, 12, 52],
         'Address': ['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
         'Qualification': ['MCA', 'Phd', 'Bcom', 'B.hons'],
         'Salary': [1000, 2000, 3000, 4000]}

df1 = pd.DataFrame(data1, index=[0, 1, 2, 3])

df2 = pd.DataFrame(data2, index=[2, 3, 6, 7])

print(df1, "\n\n", df2)

result = pd.concat([df1, df2], axis=1, join='inner')
result
```

### Joining DataFrame
* The .join() method in Pandas is used to combine columns of two DataFrames based on their indexes.
#### Joining DataFrames Using .join()
* If both DataFrames have the same index, we can use the .join() function to combine their columns. This method is useful when we want to merge DataFrames based on their row indexes rather than columns.

```python
import pandas as pd


data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32]}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}

df1 = pd.DataFrame(data1,index=['K0', 'K1', 'K2', 'K3'])

df2 = pd.DataFrame(data2, index=['K0', 'K2', 'K3', 'K4'])

print(df1, "\n\n", df2)
```

# Slicing Subsets of Rows and Columns 

We can select specific ranges of our data in both the row and column directions using either label or integer-based indexing.

* loc is primarily label based indexing. Integers may be used but they are interpreted as a label.
* iloc is primarily integer based indexing

To select a subset of rows and columns from our DataFrame, we can use the iloc method. 
For example, we can select B and C (columns 1 and 2 if we start counting at 1)

```python
df
```

```python
# iloc[row slicing, column slicing]
df.iloc[0:2, 0:3]  # Slicing row 1 to row 2 & column 1 to cloumn 3 (column 1, column 2 & column 3)
```

```python
df.iloc[1:4, 1:3] # Counting starts from 0, so here row 2 & column 2
```

Indexing by labels loc differs from indexing by integers iloc. With loc, the both start bound and the stop bound are inclusive. When using loc, integers can be used, but the integers refer to the index label and not the position.

For example, using loc and select 1:2 will get a different result than using iloc to select rows 1:2

```python
# Selecting by row  label
df.loc['2' : '3']  # selecting row 2 and row 3
```

```python
# Selecting by column label
df.loc[:,['A','C']] 
```

```python
# Specific row & column are selected by specifying row & column
df.loc[['1', '2','3'],['B','C']]
```

### Working with Missing Data in Pandas
* missing data occurs when some values are missing or not collected properly and these missing values are represented as:

* None: A Python object used to represent missing values in object-type arrays.
* NaN: A special floating-point value from NumPy which is recognized by all systems that use IEEE floating-point standards.

```python
# Checking Missing Values in Pandas
# Using isnull()
# isnull() returns a DataFrame of Boolean value where True represents missing data (NaN).
import pandas as pd
import numpy as np

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

mv = df.isnull()

print(mv)
```

```python
# Using isna()
# isna() returns a DataFrame of Boolean values where True indicates missing data (NaN)
import pandas as pd
import numpy as np

data = {'Name': ['Amit', 'Sita', np.nan, 'Raj'],
        'Age': [25, np.nan, 22, 28]}

df = pd.DataFrame(data)

# Check for missing values using isna()
print(df.isna())
```

### Filling Missing Values in Pandas

```python
# Using fillna()
# fillna() used to replace missing values (NaN) with a given value.
# Fill Missing Values with Zero:

import pandas as pd
import numpy as np

d = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(d)

df.fillna(0) # Fill Missing Values with Zero
```

```python
# Fill with Previous Value (Forward Fill)

#  The pad method is used to fill missing values with the previous value.
df.fillna(method='pad')

# df.fillna(method='ffill') # This can also be used for forward fill
```

```python
# Fill with Next Value (Backward Fill)
# The bfill function is used to fill it with the next value.
df.fillna(method='bfill')
```

```python
# Dropping Missing Values in Pandas
# The dropna() function used to removes rows or columns with NaN values.
```

```python
# Dropping Rows with At Least One Null Value
# Remove rows that contain at least one missing value.

import pandas as pd
import numpy as np

dict = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, np.nan, 45, 56],
        'Third Score': [52, 40, 80, 98],
        'Fourth Score': [np.nan, np.nan, np.nan, 65]}
df = pd.DataFrame(dict)

df.dropna()
```

```python
# Dropping Columns with At Least One Null Value

# To remove columns that contain at least one missing value we use dropna(axis=1).
dict = {'First Score': [100, np.nan, np.nan, 95],
        'Second Score': [30, np.nan, 45, 56],
        'Third Score': [52, np.nan, 80, 98],
        'Fourth Score': [60, 67, 68, 65]}
df = pd.DataFrame(dict)

df.dropna(axis=1)
```

```python
# Pandas dataframe.drop_duplicates()
# The drop_duplicates() method in Pandas is designed to remove duplicate rows from a DataFrame based on all columns or specific ones.
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Alice", "David"],
    "Age": [25, 30, 25, 40],
    "City": ["NY", "LA", "NY", "Chicago"]
}

df = pd.DataFrame(data)  # converting to dataframe

print("Original DataFrame:")
print(df)

df_cleaned = df.drop_duplicates()

print("\nModified DataFrame (no duplicates)")
print(df_cleaned)
```
