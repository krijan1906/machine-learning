# Numpy

* NumPy is the fundamental package for scientific computing in python.
* NumPy is a python libraries that provides a multidimensional array object, various derived objects.
  
* NumPy used for fast operation of an arrays, includes mathematical, logical, shape manipulation, sorting , selecting , basic linear algebra, statistical operations, random simulation.

* NumPy arrays facilate advance mathematical operations on large numbers of data.

* The major benefit of using Numpy as compared to regular python list is that numpy arrays are space & time efficient & can do the same task much more quickly.

* Numpy array consumes less memory than python list.
* Numpy array is fast as compared to the python list.
* List: [1, 2, 3, 4]
* Array: [1 2 3 4]

```python
x1 = np.array([1,2,3,4])
print(x1)
print("Data Type:", x1.dtype)

x2 = np.float32(x1)  # converting to float
print(x2)
print("Data Type:", x2.dtype)

x3 = np.int_(x2)  # converting to int
print(x2)
print("Data Type:", x3.dtype)

x4 = x3.astype(float)  # converting to float
print(x4)
print("Data Type:", x4.dtype)
```

### Arithmetic Operation in Numpy Array:


```python
import numpy as np
var = np.array([1,2,3,4])
add = var + 3 # Add each number by 3
print(add) 
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])
add = var1 + var2
print(add)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([1,2,3,4])
add = np.add(var1,var2)
print(add)
```

```python
var1 = np.array([1,2,3,4])
sub = var1 - 3
print(sub)
```

```python
var1 = np.array([1,2,3,4])
sub = np.subtract(var1, 3)
print(sub)
```

```python
var1 = np.array([1,2,3,4])
mul = var1 * 3
print(mul)
```

```python
var1 = np.array([1,2,3,4])
div = var1 / 3
print(div)
```

```python
var1 = np.array([1,2,3,4])
mod = var1 % 3  # modulus
print(mod)
```

```python
var1 = np.array([1,2,3,4])
reci = np.reciprocal(var1)
print(reci)
```

```python
# 2D array
var21 = np.array([[1,2,3,4],[1,2,3,4]])
var22 = np.array([[1,2,3,4],[1,2,3,4]])
add = var21 + var22
print(add)
```

```python
# Arithmetic Functions
import numpy as np
x = np.array([3,2,1,4,5,3,2])
print("Min:", np.min(x))
print("Max:", np.max(x))
print("Min Position:", np.argmin(x)) # return the position of minimum value in array
print("Max Position:", np.argmax(x)) # return the position of maximum value in array
print("Square Root:", np.sqrt(x))
print("Sin:", np.sin(x))
print("Cos:", np.cos(x))
print("cumsum:", np.cumsum(x)) # add previous number to current value
```

### Indexing Numpy Array:

```python
import numpy as np
x = np.array([9,8,7,6]) # Indexing starts from 0
print(x[1])
print(x[-1]) # Indexing starts from -1 from last
print(x[-3]) 
```

```python
# 2D array
x = np.array([[9,8,7], [4,5,6]])
print(x)
print("Dimension:", x.ndim)
print(x[0,2])  # row 0 and column 2
```

```python
# 3D array
x = np.array([[
    [1,2],
    [6,7]
]])
print(x)
print("Dimension:", x.ndim)
print(x[0, 1, 1])
```

```python
# Slicing array
# Syntax: [start:stop:step]

import numpy as np
x = np.array([1,2,3,4,5,6,7])
print(x)
print()
# Slicing
print(x[1:5]) # Slicing from 2 to 5
print(x[1:]) # Slicing from 2 to end
print(x[:5]) # Slicing from start to 5
print(x[::2]) # Slicing from start to end with step size 2
```

```python
x = np.array([[1,2,3,4,5], [9,8,7,6,5],[11,12,13,14,15]])
print(x)
print(x[1,1:]) # slicing from 8 to 5
```

```python
# Iteration
# Iterating Numpy Array
import numpy as np
x = np.array([9,8,7,6,5,4,3])
for i in x:
    print(i)
```

```python
# Iteration in 2D array
x = np.array([[1,2,3,4],[1,2,3,4]])
print(x)

for j in x:
    print(j)

print()
# Iteration for 2D array
for k in x:
    print(k)
    print()
    for l in k:
        print(l)
```

```python
# Iteration in 3D array
x = np.array([[[1,2,3,4],[1,2,3,4]]])
for i in x:
    print(i)
    print()
    for j in i:
        print(j)
        print()
        for k in j:
            print(k)
```

```python
# We can also use nditer() function instead of using multiple for loop
x = np.array([[[1,2,3,4],[1,2,3,4]]])
for i in np.nditer(x):
    print(i)
```

```python
# Copy vs View in Numpy Array:
# Both are used to copy

import numpy as np
x = np.array([1,2,3,4])
copy = x.copy()
x[1] = 40  # Changing original data
print("Original Data:", x)
print("Copied Data:", copy)

# Here, Original data remains unchanged.
```

```python
x = np.array([9,8,7,6,5])
view = x.view()
x[1] = 40 # Changing original data
print("Original Data:", x)
print("View Data:", view)

# Here, Original data is changed.
```

```python
# Join and Split Functions in Numpy Array:
# Join is used to merge data.
# Split is used to separate data.

import numpy as np
x = np.array([1,2,3,4])
y = np.array([9,8,7,6])
arr = np.concatenate((x,y))
print(arr)
```

```python
x = np.array([[1,2],[3,4]])
y = np.array([[9,8],[7,6]])

new_arr = np.concatenate((x,y), axis=1)  # axis =1 for column-wise merge
print(x)
print(y)
print(new_arr)
```

```python
# Split array
import numpy as np
x = np.array([1,2,3,4,5,6])
print(x)
arr = np.array_split(x,3) # split x array into 3 array
print()
print(arr)
```

```python
# Split 2D array
x = np.array([[1,2],[3,4],[5,6]])
print(x)
arr = np.array_split(x,3) # split x array into 3 array
print(arr)
```

```python
x = np.array([[1,2],[3,4],[5,6]])
print(x)
arr = np.array_split(x,3, axis=1) # split x array into 3 array along column-wise
print(arr)
```

```python
# Search
import numpy as np
x = np.array([1,2,3,4,2,5,2,5,6,7])
y = np.where(x == 2) # Position or index number of 2 in array
print(y)
```

```python
x = np.array([4,2,3,4,2,5,2,5,6,7])
y = np.where(x%2 == 0) # Position or index number of 2 in array
print(y)
```

```python
# Search Sorted Array:
# Performs a binary search in the array, and returns the index where the specified value would be inserted to maintain the search order.
x = np.array([1,2,3,4,6,7,8])
y = np.searchsorted(x, 5) # Position of 5 to be sorted in array from left-side to right
print(y)
```

```python
x = np.array([1,2,3,4,8,9,10])
y = np.searchsorted(x, 5, side="right") # Position of 5 to be sorted in array towards right-side
print(y)
```

```python
x = np.array([1,2,3,4,8,9,10])
y = np.searchsorted(x, [5,6,7], side="right") # Position of 5 to be sorted in array towards right-side
print(y)
```

```python
# Sort Array: Sort the given array in order corresponding to elements like numberic or alphabetical in ascending or descending order 
x = np.array([[4,2,3], [1,12,5], [22,52,6]])
print(np.sort(x))
```

```python
# Sort in alphabetical order
x = np.array(["a", "s", "d", "f"])
print(np.sort(x))
```

```python
# Sorting
x = np.array([[23, 1, 345, 2, 576, 234, 14], [11, 34, 253, 256, 21, 67, 72]])
np.sort(x, axis = 1)  # column-wise sorting
```

```python
#np.sort(x)    
np.sort(x, axis = 0) # row-wise sorting
```

```python
np.sort(x, kind = 'mergesort') # Merge sort divide the array into half & then compare & sort at correct position
```

```python
np.sort(x, kind = 'quicksort') # Quicksort take a pivot element, compare each element with that pivot & then sort at correct position
```

```python
# Filter Array: Getting some elements out of an existing array and creating a new array out of them.
x = np.array(["a", "s", "d", "f"])
f = [True, False, False, True]
new_arr = x[f] # return only True value
print(new_arr) 
print(type(new_arr))
```

```python
# Resize --> Change the size of array
x = np.array([1,2,3,4,5,6]) # 1D array
y = np.resize(x, (2,3))  # 2D array with 2 row and 3 columns
print(y)
```

```python
# flatten --> Convert into 1D array
x = np.array([1,2,3,4,5,6]) # 1D array
y = np.resize(x, (2,3))  # 2D array with 2 row and 3 columns
print(y)
print()
print(y.flatten())  # Convert into 1D array
```

```python
# ravel --> Convert into 1D array
x = np.array([1,2,3,4,5,6]) # 1D array
y = np.resize(x, (2,3))  # 2D array with 2 row and 3 columns
print(y)
print()
print(y.ravel())  # convert into 1D array
```

```python
# 2D array Operation:

x =np.array([[1,2],[3,4]])
y =np.array([[3,4],[4,5]])
x + y # Addition of two array
```

```python
# Dot Product ---> For matrix multiplication
# Number of rows in 1st matrix must be equal to number of columns in 2nd matrix.
print(x)
print(y)
np.dot(x,y)  # eg. 1st entry in 1st row * 1st entry in 1st column + 2nd entry in 1st row * 2nd entry in 2nd column +...
```

```python
# Array multiplication
x = np.array([[1,2,3],[1,2,3]])
print(x*x)
```

```python
# Cross Product
x =np.array([[1,2],[3,4]])
y =np.array([[3,4],[4,5]])
np.cross(x,y) # Use matrix determinant rule; 1*4 - 3*2, 3*5 - 4*4
```

```python
# Matrix function in Numpy array:
# Transpose
import numpy as np
x = np.matrix([[1,2,3], [4,5,6]])
print(x)
print()
print(np.transpose(x))
```

```python
# swapaxes
x = np.matrix([[1,2,3], [4,5,6]])
print(x)
print()
print(np.swapaxes(x, 0,1)) # 0 axis changes to 1 axis
```

```python
# Inverse
x = np.matrix([[1,2], [3,4]])
print(x)
print()
print(np.linalg.inv(x))
```

```python
# Power function
x = np.matrix([[1,2], [3,4]])
print(x)
print()
print(np.linalg.matrix_power(x,2)) # power by 2
```

```python
# det function --> Finds the determinant
x = np.matrix([[1,2], [3,4]])
print(x)
print()
print(np.linalg.det(x)) 
```

# Array

```python
# Array is collection of elements of similar datatypes
# Create an array containing fruits names:

fruits = ["Orange", "Mango", "Banana"]
fruits
```

An array is a special variable, which can hold more than one value under a single name & access the values by referring to an index number.

```python
# Accessing the element from an array
fruits[1]
```

```python
# Update/Modify the value of an array
fruits[2] = "Apple"
fruits
```

```python
# Length of array
len(fruits)
```

```python
# Looping an array
for x in fruits:   # Print each items of an array, here x is variable that represents each elements from fruits array name list
 print(x)
```

```python
# Append
fruits.append("Pineapple") # Add items to an array at last positon
fruits
```

```python
# Insert
fruits.insert(1,"Watermelon") # Insert the values at index 1
fruits
```

```python
# Deleting an element from an array
fruits.pop(1) # Index 1 array item is deleted
fruits
```

```python
# Remove an element from an array
fruits.remove("Orange")
fruits
```

```python
cars = ["Ford", "Volvo", "BMW"]

cars.remove("Volvo")

print(cars)
```
