### SciKit-Learn
* Scikit-learn is an open-source library that simplifies the process of building machine learning models.
* Focus on modeling data
 * It features various classification, regression and clustering algorithms including support vector machines, random forests, gradient boosting, k-means and DBSCAN.
 * Makes model building fast and reliable
 * Provides ready-to-use tools for training and evaluation
 * Reduces complexity by avoiding manual implementation of algorithms

#### Steps to build a model with Scikit-learn:
* Load the dataset
* Data preprocessing
* Split data into training and testing sets
* Train a model and make predictions


##### Scikit-learn simplifies machine learning with pre-built tools and algorithms.


### Install Scikit-learn: pip install scikit-learn

### Loading a Dataset
* A dataset consists of:
    * Features (X): Input variables that describe the data
    * Target (y): The value we want to predict

* Scikit-learn provides built-in datasets like Iris, Digits and Boston Housing.
#### Using the Iris dataset:
* load_iris() loads the data


```python
from sklearn import datasets # importing dataset from sklearn
```

```python
ds = datasets.load_iris() # loading the iris from datasets; as it is a supervised learning model
```

```python
ds
```

```python
feature_names = ds.feature_names 
target_names = ds.target_names 
  
print("Feature names:", feature_names) 
print("Target names:", target_names) 
```

```python
# X stores feature data
# y stores target labels
X = ds.data 
y = ds.target 
```

#### Splitting the Dataset
* To evaluate a model fairly, we split data into:
    * Training set: Used to train the model
    * Testing set: Used to evaluate how well the model generalizes

#### After splitting, we get:
     X_train, y_train -> Training data
     X_test, y_test -> Testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Checking the shapes
print("X_train Shape:",  X_train.shape)
print("X_test Shape:", X_test.shape)
print("Y_train Shape:", y_train.shape)
print("Y_test Shape:", y_test.shape)
```

```python
from sklearn.svm import SVC # SVC is support vector classifier and svm is support vector machine
# Define the model
model = SVC() # Using support vector classifier in model, sets the model

# Use the model and fit it to data(training data) and target
model.fit(X_train, y_train) 

print(model) # printing additional info about a model
```

```python
# Make Predictions
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn import metrics 
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python


#Sumarized the fitted of the model
print (metrics.classification_report(y_test, y_pred)) 
```

we get 98% accuracy.
