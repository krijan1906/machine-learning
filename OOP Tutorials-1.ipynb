### Object Oriented Programming
* Common topic in almost all popular programming languages like Python, C++, Java
* Common idea but with different syntax
* Object-Oriented Programming (OOP) is a programming language model organized around objects rather than "actions" and data rather than logic.
* OOPs is a way of organizing code that uses objects and classes to represent real-world entities and their behavior.


#### OPP is just a style/way to write a code
* OPP does not enhances the functionality of code but it changes the writing style of code
* OPP manages the code in better way

### Concepts of OOP
    * Encapsulation
    * Abstraction
    * Inheritance
    * Polymorphism

* Encapsulation - Encapsulate data into specific class or function.

* Abstraction - Showing only useful things to user and hidding other data and function.

* Inheritance - Inheriting data members and function from parent class to child class.
* Inheritance reduces the code redundency. ie. writing the same code to different class.

* Polymorphism - (Different forms) - Function with same name but used for different purposes/tasks.

### Class:

* Classes are blueprints/layout/frameworks for creating objects.
* A class defines a set of attributes and methods that the created objects (instances) can have.

### Constructor
* Python supports a special type of method called constructor for initializing the instance variable of a class.
* A class constructor, if defined is called wheneven a program creates an object of that class.
* A constructor is called only once at the time of creating an instance.
* If two instances are created for a class, the constructor will be called once for each instance.

```python
# Constructor
class Mobile:
    def __init__(self):  # this init method is also called constructor
        print("Mobile Constructor Called")
realme = Mobile()  # creating object of class Mobile

# Constructor is called automatically while creating object of class
```

```python
# Constructor without parameter
class Mobile:
  def __init__(self):     # this  init method is also called constructor rather than method
    self.model = 'RealMe X'   # instance variable

  def show_model(self):   # instance method
    print('Model:', self.model) # accessing instance variable

realme = Mobile()   # creating object of mobile class
realme.show_model()   # Calling function
```

```python
# Constructor with parameter
class Mobile:
  def __init__(self, m):   # constructor with parameter
    self.model = m         # instance variable
realme = Mobile('Realme X')   # constructor is called when class object is created

class Mobile:
  def __init__(self, m, v=80): # m and v are formal arguments or parameters
    self.model = m
    self.volume = v
redmi = Mobile('Redmi 7s', 50)  # constructor is called when class object is created
```

```python
# Constructor
class Mobile:
  def __init__(self, m, v=80):  # m is formal argument, v=80 is default argument
    self.model = m
    self.volume = v

  def show_model(self, p):
    price = p       # local variable
    print("Model:", self.model, "and Price:", price)
    print("Volume:", self.volume)

# Passing Argument to Constructor
realme = Mobile('Realme X')   # Creating object of class, Realme X is value passed to m

# Accessing Method from outside class
realme.show_model(1000)   # 1000 is passed to p
print()
redmi = Mobile('Redmi 7s', 50)  # m = Redmi 7s, v = 50
redmi.show_model(2000)   # p = 2000

# Each time while you create an object; constructor is called
```

### Accessing Instance Variable:
#### With Instance Method
* To access instance variable, we need instance methods with self as first parameter then we can access instance variable using self.variable_name.

### Accessing instance variable inside class:
* We can access instance variable inside class using self.variable_name.

```python
class Mobile:
    def __init__(self):
        # Instance variable initialized in the constructor
        self.model = 'RealMe X'
    
    def show_model(self):    # Instance method
        print(self.model)   # Accessing the instance variable inside class

# Creating an object
realme = Mobile()

# Calling instance methods that work with instance variable
realme.show_model()        
```

### Accessing Instance Variable Outside Class:
* We can access instance variable outside class using object_name.variable_name

```python
class Mobile:
    def __init__(self):
        self.model = 'RealMe X'   # Instance variable
          
    def show_model(self):    # Instance Method
        self.model   # Accessing instance variable

realme = Mobile()  # Creating object
realme.model  # Accessing instance variable from outside class using object_name.variable_name
```

### Instance variable
* Instance variables are the variables whose separate copy is created in every object.
* If we modify the copy of instance variable in an instance, it will not effect all the copies in the other instance.

```python
# Example:
class Mobile:
    def __init__(self):      # constructor
      self.model = 'RealMe X'    # Instance variable

    def show_model(self):      # Instance method
      print(self.model)      # Accessing Instance variable inside instance method using self keyword

# Creating object of Mobile class
realme = Mobile()    # copy of model is created for realme object
redmi = Mobile()     # copy of model is created for redmi object
geek = Mobile()      # copy of model is created for geek object

# Accessing instance variable outside class using object_name.variable_name
print(realme.model)  
print(redmi.model)
print(geek.model)
print()


redmi.model = 'Redmi 7s' # Modifying instance variable value for redmi object
print(realme.model)      # instance variable value remains unchanged for realme object
print(redmi.model)   # print modified value of model for redmi object
print(geek.model)    # instance variable value remains unchanged for realme object
print()
```

### Class Variable or Static Variable
* Class variables are the variables whose single copy is available to all the instance of the class.
* If we modify the copy of class variable in an instance, it will effect all the copies in the other instance.

```python
# Example: Class Variable or Static Variable
class Mobile:
    fp = 'Yes'    # class variable
    def __init__(self):    # constructor
        self.model = 'RealMe X'  # instance variable
        
    def show_model(self):    # Instance method
        print(self.model)    # Accessing instance variable 

realme = Mobile()
realme.show_model()
```

 ### Accessing class variable inside class method:

### Namespace
* In Python, Namespace represents a memory block where names are mapped to objects.
* Class Namespace - A class maintains it's own namespace known as class namespace.
* In the class namespace, the names are mapped to class variables.

* Instance Namespace - Every instance have it's own namespace known as instace namespace.
* In the instance namespace, the names are mapped to instance variables.

```python
# Class Variable - Namespace
class Mobile:
  fp = 'Yes'    # class variable, fp points to 'Yes' called class Namespace.
 
  @classmethod  
  def is_fp(cls):       # classmethod
    print("Finger Print:", cls.fp)    # Accessing class variable

realme = Mobile()  # realme object maps to Yes, realme is instance Namespace which is mapped to 'Yes'
redmi = Mobile()  # redmi object maps to Yes, redmi is instance Namespace which is mapped to 'Yes'
geek = Mobile()  # geek object maps to Yes, geek is instance Namespace which is mapped to 'Yes'

print("Class FP:", Mobile.fp)
print("RealMe:", realme.fp)
print("Redmi:", redmi.fp)
print("Geek:", geek.fp)
print()

# Modifying value of fp to 'No'
Mobile.fp = 'No'   # Modifying class variable using Mobile class will change value of each object
print("Class FP:", Mobile.fp)
print("RealMe:", realme.fp)
print("Redmi:", redmi.fp)
print("Geek:", geek.fp)
print()
```

```python
# Example:
# Class Variable - Namespace
class Mobile:
  fp = 'Yes'       # class variable
  @classmethod     # class method
  def is_fp(cls):
    print("Finger Print:", cls.fp)    # Accessing class variable

realme = Mobile()
redmi = Mobile()
geek = Mobile()

print("Class FP:", Mobile.fp)
print("RealMe:", realme.fp)
print("Redmi:", redmi.fp)
print("Geek:", geek.fp)
print()

Mobile.fp = 'No'   # Mobile.fp = no
print("Class FP:", Mobile.fp)
print("RealMe:", realme.fp)
print("Redmi:", redmi.fp)
print("Geek:", geek.fp)
print()

realme.fp = 'Not Working'  # Modifying class variable value using realme object will modify value of realme object only.
print("Class FP:", Mobile.fp)
print("RealMe:", realme.fp)
print("Redmi:", redmi.fp)
print("Geek:", geek.fp)
print()
```

### Instance Method
* Instance methods are the methods which act upon the instance variables of the class.
* Instance method need to know the memory address of the instance which is provided through self variable by default as first parameter for the instance method.

```python
# Syntax:
'''
def method_name(self):       # Instance method without parameter/formal arguments
    function body

def method_name(self, f1,f2):   # Instance method with parameter/formal arguments
    function body
'''
```

```python
# Instance method without parameter
class Mobile:
  def show_model(self):      # Instance method
    print("RealMe X")
realme = Mobile()

class Mobile:
  def __init__(self):      # Constructor
    self.model = 'RealMe X'       # Instance variable

  def show_model(self):           # Instance method
    print(self.model)             # Accessing instance variable inside method

realme = Mobile()           # Object of Mobile class
realme.show_model()
```

```python
# Calling instance method without argument
# Instance methods are bound to obejct of the class so we call instance method with object name.
# Syntax:
'''
object_name.method_name()
Eg. realme.show_model()
'''

class Mobile:
  def show_model(self):   # Instance method
    print("RealMe X")

realme = Mobile()       # creating object
realme.show_model()     # calling instance method without argument using realme object
```

```python
# Calling instance method with argument:
# Syntax: 
'''
object_name.method_name(actual_argument)
Eg. realme.show_model(1000)
'''
```

```python
# Example:
# Instance method with parameter
class Mobile:
      def __init__(self):     # Constructor
        self.model = 'RealMe X'       # Creating Instance variable model
    
      def show_model(self, p):        #  show_model is Instance method, p is parameter
        self.price = p  # Assigning parameter p to price
        print("Model:", self.model, "Price:", self.price)  # Accessing model and price 

realme = Mobile()   # Creating object for calling show_model method
redmi = Mobile()   # Creating object for calling show_model method

# Calling instance method with argument
realme.show_model(1000)   # passing value to p using realme object
redmi.show_model(2000)    # passing value to p using redmi object
```

### Accessor Method or Getter Method: Part of instance method
* This method is used to access or read data of the variables. This method do not modify the data in the variable.
* This is also called as getter method.

```python
# Accessor Method or Getter Method
class Mobile:
      def __init__(self):
        self.model = 'RealMe X'       # Instance variable
    
      def get_model(self):            # Accesor method
        return self.model

realme = Mobile()
m = realme.get_model()
print(m)
```

### Mutator Method or Setter Method: Part of instance method
* This method is used to access or read and modify data of the variables. This method modify the data in the variable.
* This is also called as setter method.

```python
# Mutator Method or Setter Method without parameter
class Mobile:
      def __init__(self):
        self.model = 'RealMe X'       # Instance variable
    
      def set_model(self):            # Mutator method
        self.model = 'RealMe 2'      # setting model = RealMe 2

realme = Mobile()
# Before settting
print(realme.model)

# After setting
realme.set_model()              # calling mutator/setter method
print(realme.model)
```

```python
# Mutator Method or Setter Method wih parameter
class Mobile:
  def set_model(self, m):            # Mutator method
    self.model = m

realme = Mobile()
realme.set_model('RealMe X')              # calling mutator method, RealMe X is value for parameter m
print(realme.model)
```
