### Files
File is the collection of data that is available to a program. We can retrieve and use data stored in a file whenever we required.

### Types of Files
* Text file - It stores data in the form of characters. It is used to store characters and strings.
* Binary file - It stores data in the form of bytes, a group of 8 bits each. It is used to store text, images, pdf, csv, video and audio.

### File Handling 
File handling is the process of creating, opening, reading, writing, appending, and closing files so that data can be stored permanently and accessed later.

In simple words, file handling allows a program to interact with files stored on the computer instead of working only with temporary data in memory.

### Why is File Handling Needed?
* To store data permanently unless someone remove it.
* Stored data can be shared.
* To read data from files (e.g., datasets, logs, reports)
* To update or modify files
* To manage large amounts of data

### Common File Modes
* Mode ----->  Meaning
* r	--------> Read (default)
* w	--------> Write (creates/overwrites file)
* a	--------> Append
* x	--------> Create new file if not and show error FileExistError if the specified file is available
* b	--------> Binary mode
* t	--------> Text mode (default)
* r+ -------> Read and write
* w+ -------> Write then read, it overwrite existing data
* a+ -------> Append then read, this doesn't overwrite existing data

### File Methods
* Method ------> Description
* read() ------> Reads entire file
* readline() --> Reads one line
* readlines()--> Reads all lines into list
* write() -----> Writes string
* writelines()-> Writes list of strings
* close() -----> Closes file

```python
# Opening a file
# If we want to use a file or its data, first we have to open it.
# open() - used to open a file. It returns a pointer to the begining of the file.
# close() - used to close the opened file.
# Syntax: open('filename', mode='r', encoding='utf-8')
# filename used to represent a name of a file.
# mode represents the purpose of opening the file. It defaults to 'r' which means open for reading in text mode.


f = open('student.txt', 'w') # create student.txt file 
print(f)

# Here, f is file handler/file object, student.txt is file name and w is file mode.
# We can also give file path instead of file name.
```

```python
# writelines() - used to store/write group of string(list, tuple, set) into the file represented by the file object.
f = open('student.txt', mode='w') # Open student.txt in write mode
lst = ['Rahul\n', 'Sonam\n', 'Sumit\n', 'Rani\n', 'Raj']
f.writelines(lst)  # read all lines into the list
print(lst)
f.close()
print('Success')
```

```python
# Reading data from file
# read(size) - used to read data/content from a file and returns it as string in text mode or bytes object in binary mode.
# where size represents the number of bytes to be read from the begining of the file.
f = open('student.txt', mode='r')
data1 = f.read(2)  # read 2 characters
data2 = f.read(3)  # read next 3 characters
print(data1)
print(data2)
f.close()
print('Completed!!!')
```

```python
# Reading data from file
# readline() - This method is used to read single line from a file.
# readlines() -This method is used to read all lines from a file. It will return list of line.

f = open('student.txt', mode='r')
data1 = f.readline()
data2 = f.readlines()
print(data1)
print(data2)
f.close()
```

```python
f = open('student.txt', mode='r')
data = f.readlines()
for i in data:
  print(i, end='')
f.close()
```

```python
# tell() and seek() method
# tell() - This method is used to find current position of file pointer from begining of the file. Position starts from 0.
f = open('student.txt', mode='r')
print(f.tell())  # return the current position of file pointer
data1 = f.read(5) # read 5 character from current position
print(data1)

print(f.tell())
data2 = f.read(3)  # read next 3 characters
print(data2)
print(f.tell())
```

```python
# seek() - This method is used to move file pointer from one position to another position from begining of the file. Position starts from 0 and it must be positive integer.
f = open('student.txt', mode='r')
print(f.tell()) # read the current position of pointer
f.seek(7)  # move file pointer to 7th position from 0
print(f.tell())
f.seek(2)  # move file pointer to 2nd position
print(f.tell()) # return the current position of pointer
```

### Pickling 
* Pickling is a process of converting a class object into a byte stream so that it can be stored into a file.
* This is also called as object serialization.
* We use pickle module to perform pickling and unpickling.
* dump() function is used to perform the pickling. It returns the pickled representation of the object as a bytes object, instead of writing it to file. This method belongs to pickle module.

* Unpickling is a process whereby byte stream is converted back into a class object. It is inverse operation of pickling. This is also called as de-serialization.
* Pickling and unpickling should be done using binary files since they support byte streams.
* We use pickle module to perform pickling and unpickling.
* load() function is used to read an pickled object from a binary file and returns it into object. This method belongs to pickle module.

#### Why do we need Pickling and Unpickling
* When we store some structured data in the file and want to perform calculation that time we need pickling and unpickling.

```python
class Student:
   def __init__(self, name, roll, address):   # Constructor
     self.name = name   
     self.roll = roll
     self.address = address

   def disp(self):
     print(f'Name:{self.name} Roll:{self.roll} Address:{self.address}')

# You can also Save this file as std.py
```

```python
import pickle, std  # importing std.py file
# Pickling
with  open('student.dat', mode='wb') as f:  # write in binary mode
  std1 = std.Student('Rahul', 101, 'Ranchi') # creating object
  std2 = std.Student('Sonam', 102, 'Dhanbad') # creating next object
  pickle.dump(std1, f)  # convert to byte string
  pickle.dump(std2, f)
  print('Pickling Done!!!')
```

```python
# Unpickling
with open('student.dat', mode='rb') as f: # read in binary mode
  obj1 = pickle.load(f)
  obj2 = pickle.load(f)
  print('Unpickling Done!!!')
  obj1.disp()
  obj2.disp()
```

```python
# Separating pickling and unpickling 

# Pickling
# Save this file as pick.py
# Storing object in the file
import pickle, std
n = int(input('Enter Number of Students: '))
with  open('student.dat', mode='wb') as f:
  for i in range(n):
    name = input('Enter Student Name: ')
    roll = int(input('Enter Roll: '))
    address = input('Enter Address: ')
    std1 = std.Student('name', roll, address)
    pickle.dump(std1, f)

print('Pickling Done!!!')
```

```python
# Unpickling
# Save this file as unpick.py
# Reading object from the file
import pickle, std
with open('student.dat', mode='rb') as f:
  while True:
    try:
      obj = pickle.load(f)
      obj.disp()
    except EOFError:
      print('Done')
      break
      
# To run this first run pick.py file
  
  
```

### Directory 
os module - This module is used to perform simple operation on directories.This module represents operating system dependent functionality.

* getcwd() - This method is used to know the currently working directory
* mkdir('dirname') - This method is used to create a directory in the present directory.
* makedirs('parentdir/child/grandchilddir') - This method is used to recursively create sub directories.
* chdir('dirname') - This method is used to change current working directory.
* rename('oldname', 'newname') - This method is used to change the directory name.
* rmdir('dirname') - This method is used to remove a directory from the current working directory. We can also specify path for directory.
* removedirs('dirname') - This method is used to recursively remove all directories.
* walk() - This method is used to know contents of a directory including sub directory. It returns an iterator object whose contents can be displayed using for loop.
* The iterator object contains directory path, directoryname, filename found in the specified directory.

```python
import os
print(os.getcwd())
```
